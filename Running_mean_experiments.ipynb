```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N_data_points = 300
tau = 100
t = np.arange(N_data_points)
y = np.exp(-2*t/tau)
noise = np.random.normal(0,0.5,N_data_points)
#y2 = y + noise
y2 = np.random.normal(y,0.1,N_data_points)
#print(t)
#print(noise)
#print(y)
#print(y2)
```

```python
a = 0.08 # a = 1/tau
print('a: ', a)
x_mean = [] #np.zeros(N_data_points)
for i,x in enumerate(y2):
    #print(i,x)
    if i == 0:
        x_mean.append(x)
    #print(x_mean)
    if i < N_data_points-1:
        x_mean.append((1 - a)*x_mean[i] + a*x)
#print(x_mean)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.plot(t,y)

plt.subplot(1,3,2)
plt.plot(x_mean, 'darkred')
plt.scatter(t,y2, color = 'plum')
plt.plot(t,y, 'darkgreen')

plt.subplot(1,3,3)
plt.plot(x_mean, 'r')
plt.plot(t,y)
plt.show()
```
